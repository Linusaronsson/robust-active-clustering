```python
from rac.experiment_data import ExperimentReader
metrics = ["rand", "ami"]
ac = ExperimentReader(metrics=metrics)

config = {
    "general_options": {
        "experiment_name": "infer_exp",
        "num_repeats": 10,
        "n_workers": 10,
        "local": True,
        "verbose": False
    },
    "experiment_options": {
        "seed": 33,
        "num_feedback": 1,
        "noise_level": [0.0],
        "persistent_noise_level": 0.0,
        "force_global_update": True,
        "save_matrix_data": False,
        "infer_sims2": True,
    },
    "sim_init_options": {
        "K_init": 1,
        "sim_init": 0.1,
        "sim_init_type": ["uniform_random", "random_clustering"],
        "init_noise_level": 0
    },
    "query_strategy_options": {
        "acq_fn": ["freq", "QECC", "COBRAS"],
        "local_regions": ["pairs", "none"]
    },
    "dataset_options": {
        "dataset": "synthetic",
        "n_samples": 500,
        "n_clusters": 10,
        "class_balance": None
    }
}

data = ac.read_all_data(folder="../experiment_results_local/infer_exp")
ac.generate_AL_curves(
    data,
    save_location="../plots/infer_exp",
    categorize=["dataset", "noise_level"],
    compare=["acq_fn", "local_regions"], 
    vary=["x"],
    auc=True,
    summary_method="auc_max_ind",
    indices=[], 
    threshold=1,
    err_style="band",
    marker="o",
    markersize=6,
    capsize=6,
    linestyle="solid",
    **config
)
```
