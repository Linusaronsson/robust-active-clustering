```python
import numpy as np
import itertools
def num_edges(cluster_sizes):
    num_pos = 0
    num_neg = 0
    k = len(cluster_sizes)
    clusts = np.arange(k)
    for k1, k2 in itertools.combinations(clusts, 2):
        num_neg = cluster_sizes[k1] * cluster_sizes[k2]

    for k1 in clusts:
        num_pos += cluster_sizes[k1] * (cluster_sizes[k1] - 1) / 2
    return num_neg, num_pos
```
